#### **2. Data Cleaning**

> Setelah melakukan eksplorasi data sebelumnya, terdapat beberapa langkah yang dapat dilakukan dalam proses pembersihan data *(data cleaning)*:
> - Perbaikan penamaan kolom yang mengandung *whitespace* yang tidak perlu.
> - Perbaikan penamaan list dari isi kolom yang mengandung *whitespace* yang tidak perlu.
> - Penghapusan karakter-karakter yang tidak perlu.
> - Penyesuaian tipe data pada beberapa kolom agar sesuai dengan jenis data yang sebenarnya.

##### a. Manipulasi kolom meliputi penggantian nama kolom, penghapusan kolom, atau penambahan kolom

````python
# Mengganti penamaan kolom yang mengandung spasi yang tidak perlu

df = df.rename(columns={' Product ': 'Product',
                        ' Discount Band ': 'Discount Band',
                        ' Units Sold ': 'Units Sold',
                        ' Manufacturing Price ': 'Manufacturing Price',
                        ' Sale Price ': 'Sale Price',
                        ' Gross Sales ': 'Gross Sales',
                        ' Discounts ': 'Discounts',
                        '  Sales ': 'Sales',
                        ' COGS ': 'COGS',
                        ' Profit ': 'Profit',
                        ' Month Name ': 'Month Name',
                        })

# Menampilkan nama kolom setelah perubahan
df.columns
````

Perbaikan nama kolom dapat dilakukan dengan metode `rename`. Dalam tahap ini, dilakukan perbaikan penamaan kolom yang sebelumnya mengandung *whitepace* yang tidak diperlukan. Setelah dilakukan perubahan, nama-nama kolom akan terlihat lebih rapi dan sesuai dengan standar penamaan yang lebih baik.

Pada tahap ini,  tidak diperlukan penghapusan kolom karena tidak terdapat data yang duplikat yang perlu dihilangkan. Selain itu, juga tidak perlu dilakukan penambahan kolom karena tidak ada informasi tambahan yang perlu ditambahkan ke dalam dataset. Tahap ini fokus pada perbaikan penamaan kolom.

##### b. Penggantian tipe data yang sesuai dengan seharusnya. Untuk nilai harga dengan simbol mata uang, simbol tersebut dapat dibuang supaya dapat terkonversi ke tipe data yang seharusnya. Perhatikan format penulisan angkanya apakah ada tanda yang membuat angka hanya bisa bertipe data string.
> (Tambahan: Di suatu kolom yang bernilai uang, terpadat beberapa data yang mengandung karakter '(' dan ')' di dalamnya sehingga harus ditake-out karakter tersebut.)

##### c. Cek dan handling missing value. (Hint: terdapat non-standard missing value pada kolom yang semestinya bertipe data integer, handling dapat dilakukan dengan imputasi dengan angka nol '0')

````python
# Menghapus spasi yang tidak diperlukan pada isi list kolom 

df['Product'] = df['Product'].str.replace(' ', '')
df['Discount Band'] = df['Discount Band'].str.replace(' ', '')
````

Dari analisis sebelumnya, didapatkan bahwa dalam kolom `Product` dan `Discount Band`, terdapat spasi yang tidak diperlukan pada nama item di dalam list. Untuk memperbaiki penamaan dari isi list tersebut, dapat menggunakan metode `str.replace`. Hal ini dapat membantu memperbaiki penamaan sehingga tidak terdapat spasi yang tidak diperlukan.

````python
# Menyimpan nama-nama kolom pada sebuah variabel
cleanColumns = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']

# Melakukan looping pada cleanColumns untuk mengubah data
for col in cleanColumns:
    # Menghapus seluruh karakter maupun spasi selain digit dan desimal
    df[col] = df[col].str.replace(r'[^\d.]+', '', regex=True)
    # Mengisi data yang kosong (missing values) menjadi angka 0
    df[col] = df[col].replace('', '0')
````

Untuk mengubah tipe data ke format yang sesuai, langkah pertama adalah membersihkan karakter-karakter yang tidak diperlukan, seperti simbol uang ($), koma (,), tanda strip (-), dan lainnya. Langkah ini dilakukan dengan menggunakan metode `str.replace`.

Setelah membersihkan karakter-karakter tersebut, beberapa kolom dapat menjadi kosong *(null)* karena sebelumnya hanya berisi karakter. Untuk menghindari nilai null, langkah selanjutnya adalah mengganti nilai-nilai kosong tersebut dengan angka '0'.

````python
# Mengubah tipe data menjadi datetime

df['Date'] = pd.to_datetime(df['Date'])
# df['Month Number'] = pd.to_datetime(df['Month Number'], format='%m').dt.month
````

Setelah membersihkan karakter dan menghindari nilai null, selanjutnya adalah mengubah tipe data `Date` menjadi datetime dengan metode `pd.to_datetime`.

````python
# Mengubah tipe data menjadi float

df['Units Sold'] = df['Units Sold'].astype(float)
df['Manufacturing Price'] = df['Manufacturing Price'].astype(float)
df['Sale Price'] = df['Sale Price'].astype(float)
df['Gross Sales'] = df['Gross Sales'].astype(float)
df['Discounts'] = df['Discounts'].astype(float)
df['Sales'] = df['Sales'].astype(float)
df['COGS'] = df['COGS'].astype(float)
df['Profit'] = df['Profit'].astype(float)
````

Setelah itu, data yang berisi angka diubah dari `Object` menjadi `Float` dengan metode `astype`.

#### **3. Data Normalization**

##### a. Lakukan normalisasi data yang sudah diolah sebelumnya sampai dengan 3NF.

##### b. Jumlah tabel setelah dinormalisasi sebanyak 5 tabel termasuk tabel utama.

##### c. Perhatikan penggunaan primary key dan foreign key antar tabel.

Dalam melakukan normalisasi database, langkah pertama yang dilakukan adalah memastikan bahwa data sudah memenuhi aturan 1NF *(First Normal Form)*, dimana setiap kolom hanya berisi satu nilai. Setelah memastikan hal tersebut, langkah selanjutnya adalah melakukan normalisasi ke tingkat 2NF *(Second Normal Form)* dan 3NF *(Third Normal Form)*.

Dari data yang telah diperiksa, dapat disimpulkan bahwa dataframe sudah memenuhi aturan 1NF karena setiap kolom hanya berisi satu nilai. Oleh karena itu, langkah selanjutnya adalah melakukan normalisasi ke tingkat 2NF dan 3NF.

````python
# Membuat tabel segment yang berisi isi list dan menyimpannya ke variabel

segments = pd.DataFrame(df['Segment'].unique())
segments
````

````python
# Mengganti nama isi segment menjadi index

i = 0 

# Melakukan proses looping untuk mengakses seluruh baris
for mainName in df['Segment']:
    for listName in segments.index:
        if mainName == segments.iloc[listName][0]:
            df.loc[i, 'Segment'] = listName + 1
    i += 1

df
````

````python
# Membuat tabel countries yang berisi isi list dan menyimpannya ke variabel

countries = pd.DataFrame(df['Country'].unique())
countries
````

````python
# Mengganti nama isi country menjadi index

i = 0

# Melakukan proses looping untuk mengakses seluruh baris
for mainName in df['Country']:
    for listName in countries.index:
        if mainName == countries.iloc[listName][0]:
            df.loc[i, 'Country'] = listName + 1
    i += 1

df
````

````python
# Membuat tabel products yang berisi isi list dan menyimpannya ke variabel

products = pd.DataFrame(df['Product'].unique())
products
````

````python
# Mengganti nama isi product menjadi index

i = 0

# Melakukan proses looping untuk mengakses seluruh baris
for mainName in df['Product']:
    for listName in products.index:
        if mainName == products.iloc[listName][0]:
            df.loc[i, 'Product'] = listName + 1
    i += 1
            
df
````

````python
# Membuat tabel discount bands yang berisi isi list dan menyimpannya ke variabel

discountBands = pd.DataFrame(df['Discount Band'].unique())
discountBands
````

````python
# Mengganti nama isi discount band menjadi index

i = 0

# Melakukan proses looping untuk mengakses seluruh baris
for mainName in df['Discount Band']:
    for listName in discountBands.index:
        if mainName == discountBands.iloc[listName][0]:
            df.loc[i, 'Discount Band'] = listName + 1
    i += 1

df
````

Pada tahap ini, tabel-tabel baru dibuat, yaitu tabel segments, countries, products, dan discount bands yang berisi list nama dan indeksnya. Indeks ditambahkan dengan angka 1 karena di Visual Studio Code, indeks dimulai dari 0 sementara di PostgreSQL, indeks dimulai dari angka 1. Karena data ini akan diolah ke PostgreSQL selanjutnya, maka indeksnya ditambahkan 1.

Setelah membuat tabel-tabel tersebut, nama list dari tabel utama juga diubah menjadi indeks yang mewakili masing-masing nama.

````python
# Mengubah tipe data dari object menjadi integer

df['Segment'] = df['Segment'].astype(int)
df['Country'] = df['Country'].astype(int)
df['Product'] = df['Product'].astype(int)
df['Discount Band'] = df['Discount Band'].astype(int)
````

Dataframe di atas merupakan main table dari data yang telah dinormalisasi. Dapat lihat bahwa kolom `Segment`, `Country`, `Product`, dan `Discount Band` telah diganti menjadi angka yang merepresentasikan setiap nama berdasarkan tabel masing-masing.

##### 1. a. Berikan tabel yang berisikan informasi total profit yang didapatkan di tiap jenis segmentasi. Jangan ambil data yang tidak diskon.

```sql
SELECT a.name as "Segment",
	   ROUND(SUM(b.profit)::numeric, 2) as "Total Profit"
FROM segments as a
JOIN mainTable as b
ON a.id = b.segment
WHERE b.discounts > 0
GROUP BY 1
ORDER BY 2 DESC;
```

##### 1. b. Berikan tabel yang berisikan informasi rangkuman statistik yang memuat nilai rata-rata, min, dan max (dijadikan dalam kolom yang berbeda) dari Sales masing-masing negara.

```sql
SELECT a.name as "Country",
	   ROUND(AVG(b.sales)::numeric, 2) as "Average Sales",
	   MIN(b.sales) as "Minimum Sales",
	   MAX(b.sales) as "Maximum Sales"
FROM countries as a
JOIN mainTable as b
ON a.id = b.country
GROUP BY 1;
```
